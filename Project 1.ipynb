```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Data collection and processing
df=pd.read_csv("Sonar data.csv",header=None)
```

```python
# seprating data and label
x=df.drop(columns=0,axis=1)
y=df[60]
```

```python
print(x)
print(y)
```

```python
#Training and testing 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)  # 0.1 ia 10% of data to test, stratify coz we need to split the data based on rock or mine,random is to split data in particular order
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(y_train)
```

```python
print(x_train)
```

```python
# Model Training Logistic Regression
model=LogisticRegression()
```

```python
#training LR model
model.fit(x_train,y_train)
```

```python
# model evaluation 
# accuracy on training data
x_train_prediction = model.predict(x_train)
traing_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
traing_data_accuracy # number * 100 will be the accuracy
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```
