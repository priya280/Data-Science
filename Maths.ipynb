# Dot and Cross Product of vectors

```python
import numpy as np
```

```python
a=np.array([2,3])
b=np.array([4,4])
a_dot_b = np.dot(a,b)
```

```python
print(a_dot_b)
```

```python
c = np.array([40,12,45])
d = np.array([15,50,33])
c_dot_d= np.dot(c,d)
```

```python
print(c_dot_d)
```

```python
a_cross_b=np.cross(a,b)
print(a_cross_b)
```

```python
c_cross_d=np.cross(c,d)
print(c_cross_d)
```

```python
# Projection a on v
a=np.array([2,5])
v=np.array([8,-6])

#magnitude of v vector
mgv=np.sqrt(sum(v**2))
proj_of_a_on_v=(np.dot(a,v)/mgv**2)*v
print("Projection of a vector on v :",proj_of_a_on_v)
```

# Statistics

> Descriptive Statistics :
. used to describe the features of data in a study.
. provide simple summaries about sample.

1. Measure of central tendancies (Mean,median, mode)

2. Measure of variability (range, sd,variance)

> Inferential Statistics :
. Takes data and makes inferences & predictions abt larger population.

1. Sample study : represents total population

2. Observational Study : collect and analyze data.Observe correlation of data.

3. Experimental Study : conditions are controlled and manipulated by experimenter.

# Sampling Techniques : (Probability Sampling Techniques)

# 1. Simple Random Sampling -
sample is randomly picked from a larger population. eg: Employee survey in a company.

# 2. Systematic Sampling -
Sample is picked at regular intervals from population. Used for homogenous(data is equally distributed) .eg: Selecting every 10th member from a population.

# 3. Stratified Random Sampling -
Population can be subdivided into smaller grps called Strata. Samples are obtained randomly from all these stratas. eg : Smartphone states in all states.

# 4. Cluster Sampling -
Population with inherent or natural groups.It is subdivided into clusters & random clusters are taken as sample. eg: Smartphone sales randomly selected states.

# Central Tendancy : 
We use these values in data pre-processing for handling missing values.
eg: GPA of students.

> 1. Mean : if data is uniformly distributed then missing values can be replaced with mean value.

> 2. Median : if data is skewed (where there is more number of data in one side) then missing values can be replaced with median value.

> 3. Mode : if data is skewed or if there are missing values in categorical column(They contain categories or classes) then missing values can be replaced with mode value.

# Skewness :
 a. Negatively skewed data : data is more towards the right side of graph.



# Variability :

> 1. Range : rough idea abt the distribution of dataset.

> 2. Variance : how far each number in the set is from the mean.

> 3. Standard Deviation : how spread out a group of numbers is from the mean.

# Correlation  :
extent which two variables are related to each other.

> 1. Postitve Correlation
> 2. Negative Correlation

# Causation :
one event causes another event to occur.


# Hypothesis :
It is an assumption that is made based on observations of an experiment.

# Null hypothesis :
H0 is commonly accepted fact.

# Alternative Hypothesis :
Ha challenges the H0.

# Hypothesis Testing :
method carried out to test the assumtions made in the experiment.

# loss function :
helps to find distance between the predicted value and original value. we can use this for model optimization. It depends on probability.

> yi - original value of model

> yi cap - value predicted by model

> n - number of values in set

# Probability value lies between 0(impossible) & 1(certain) , 0.5(half chance)

# Probability Distribution for Random variable :
how the probabilities are distributed over the values of random variables.

# Normal Distribution or Gaussian Distribution:
arrangement of dataset(class of students) in which most of the data points(height of students) lie in middle of range(160cm to 190cm) and rest taper(160cm) off symmetrically towards either extreme(200 cm).

# Poisson Distribution : 
how many times an event is likely to occur within a specified period of time. eg : no. of accidents occuring in  a city from 6pm to 10pm.
