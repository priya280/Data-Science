```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

df = pd.read_csv('iris_data.csv')
```

```python
x = df['SepalLengthCm']
y = df['PetalLengthCm']

width=0.09
plt.bar(x, y,width,color="purple")

plt.title("Sepal vs Petal length Bar chart")

plt.ylabel('Petal Length')

plt.xlabel('Sepal Length')

 
plt.show()
```

```python

plt.hist(x, bins=25, color='yellow', edgecolor='blue',
         linestyle='--', alpha=0.5)

plt.title("Sepal vs Petal length Histogram")

plt.ylabel('Petal Length')

plt.xlabel('Sepal Length')
 
plt.show()
```

```python
plt.scatter(x, y, c=df['SepalLengthCm'], s=df['SepalWidthCm'],
            marker='D')
```

```python
KS = {'color': ['blue', 'red', 'yellow']}
sns.FacetGrid(df, hue_kws=KS, hue="Species", size=5) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
plt.show()
```

```python
df.plot(kind ="scatter",
          x ='SepalLengthCm',
          y ='PetalLengthCm')
plt.grid()
```

```python
df.drop("Id", axis=1).boxplot(by="Species", figsize=(10, 10))
plt.show()
```

```python
sns.violinplot(x="Species", y="PetalLengthCm",  palette={"blue","red","yellow"}, data=df, size=6)
plt.show()
```

```python
from matplotlib import font_manager as fm
plt.rcParams['font.size'] = 22.0
matplotlib.rcParams['text.color'] = 'w'
matplotlib.rcParams['lines.linewidth'] = 2
labels=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
explode=(0,0,0.1)
plt.pie(df, labels=labels,explode=explode, autopct='%1.1f%%', radius=1.2, colors=("y","m","b"), textprops={'fontsize': 22})
plt.title("Distribution of Sepal Length by Iris Class",  y=1.2,  fontsize=22)
plt.show()
```
