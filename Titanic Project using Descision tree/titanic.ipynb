```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("titanic.csv")
df.head()
```

```python
df.tail(10)
```

```python
df.describe()
```

```python
df.drop(['Name','SibSp','PassengerId','Ticket','Cabin','Embarked','Parch'],axis='columns',inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.groupby('Pclass').size()
```

```python
ip=df.drop("Survived",axis="columns")
target=df["Survived"]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
new_Sex=LabelEncoder()
ip["sex"]=new_Sex.fit_transform(ip['Sex'])
ip.head()
```

```python
new=ip.drop("Sex",axis='columns')
new.head()
```

```python
df.isnull()
```

```python
ip.Age[:10]
```

```python
new.Age=new.Age.fillna(new.Age.mean())
new.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(new,target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
model.predict([[3,26.000000,7.9250,0]]) #prediction if model is working
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (7,4))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
